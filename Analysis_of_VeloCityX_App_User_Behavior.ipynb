```r
cor_matrix <- cor(df[, sapply(df, is.numeric)])  # Correlation matrix for numeric columns
print(cor_matrix)  # Display the correlation matrix
```

### Correlation Analysis

The correlation matrix was created to analyze the relationships between user activities and virtual merchandise purchases. The results are summarized below:

- **Fan Challenges Completed**: A moderate positive correlation (0.159) indicates that users who engage more in fan challenges are likely to purchase more virtual merchandise.
- **Predictive Accuracy (%)**: Shows a weak positive correlation (0.022) with merchandise purchases, suggesting that accuracy in challenges does not significantly influence purchasing behavior.
- **Sponsorship Interactions**: Exhibits a negligible correlation (-0.065), implying minimal impact on merchandise purchases.
- **Time on Live 360**: There is a slight negative correlation (-0.0075), indicating no substantial effect on purchases.
- **Real-Time Chat Activity**: A weak negative correlation (-0.044) suggests limited interaction effects on purchasing behavior.

These insights suggest that while engagement in fan challenges is positively associated with merchandise purchases, other factors may require deeper analysis.

```r
library(ggplot2)
library(reshape2)

# Melt the correlation matrix
melted_cor <- melt(cor_matrix)
ggplot(data = melted_cor, aes(Var1, Var2, fill = value)) +
    geom_tile() +
    scale_fill_gradient2(low = "blue", high = "red", mid = "white", limit = c(-1, 1), name = "Correlation") +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
    ggtitle("Correlation Heatmap")
```

### Correlation Heatmap

The correlation heatmap visually represents the relationships between the various metrics. The darker shades indicate stronger correlations. 
This visualization confirms the numerical findings, particularly emphasizing the positive correlation between fan challenges completed and virtual merchandise purchases. 
The weaker correlations among other metrics highlight areas for potential improvement or further investigation in user engagement strategies.

```r
# Select relevant features
features <- df %>% select(`Fan Challenges Completed`, `Predictive Accuracy (%)`, 
                           `Virtual Merchandise Purchases`, `Sponsorship Interactions (Ad Clicks)`, 
                           `Time on Live 360 (mins)`, `Real-Time Chat Activity (Messages Sent)`)

# Scale the data
scaled_features <- scale(features)
```

```r
# Elbow Method
wss <- sapply(1:10, function(k) { kmeans(scaled_features, k, nstart = 10 )$tot.withinss })
plot(1:10, wss, type = "b", pch = 19, frame = FALSE, 
     xlab = "Number of Clusters", ylab = "Total Within Sum of Squares")

# Apply K-Means with the chosen number of clusters (let's say k = 3)
set.seed(123)  # For reproducibility
kmeans_result <- kmeans(scaled_features, centers = 3, nstart = 10)

# Add the cluster assignments to the original data
df$Cluster <- as.factor(kmeans_result$cluster)
```

```r
# Summarize the clusters
cluster_summary <- df %>% 
  group_by(Cluster) %>% 
  summarise(across(`Fan Challenges Completed`:`Real-Time Chat Activity (Messages Sent)`, mean))

print(cluster_summary)
```

We applied K-means clustering to segment users based on their engagement metrics. The analysis revealed three distinct clusters, each characterized by varying levels of fan challenges completed, predictive accuracy, and virtual merchandise purchases.

```r
# Fit the linear regression model
model <- lm(`Virtual Merchandise Purchases` ~ `Fan Challenges Completed`, data = df)

# Display the summary of the model
summary(model)
```

```r
# Load ggplot2 for visualization
library(ggplot2)

# Create a scatter plot with the regression line
ggplot(df, aes(x = `Fan Challenges Completed`, y = `Virtual Merchandise Purchases`)) +
    geom_point() +  # Scatter plot
    geom_smooth(method = "lm", col = "blue") +  # Regression line
    labs(title = "Linear Regression: Fan Challenges vs Virtual Merchandise Purchases",
         x = "Fan Challenges Completed",
         y = "Virtual Merchandise Purchases") +
    theme_minimal()
```

## Simple Linear Regression Analysis

We performed a simple linear regression to investigate the relationship between the number of fan challenges completed and virtual merchandise purchases. 

### Model Summary
- **Coefficients**:
  - Intercept: **1.996**
  - Slope (Fan Challenges Completed): **0.116**
  
- **Statistical Significance**:
  - The p-value for the slope is **0.113**, indicating that the relationship between fan challenges and virtual merchandise purchases is **not statistically significant** at the 5% significance level.

- **R-squared**: **0.0254**
  - This indicates that only **2.54%** of the variance in virtual merchandise purchases can be explained by the number of fan challenges completed.

### Visualization
The scatter plot above shows the relationship between fan challenges completed and virtual merchandise purchases, with the blue regression line indicating the expected trend.
